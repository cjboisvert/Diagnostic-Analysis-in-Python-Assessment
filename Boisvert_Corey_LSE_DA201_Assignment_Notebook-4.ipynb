#### Observations on the above plot:
This chart was difficult to interpret, so I cleaned by re-categorising the data. I did not want to omit any of the values, but I noticed that several categories were quite similar, so I created a dictionary to remap the values, ultimately lowering the number of categories by combining similar ones.

```python
# Dictionary to remap categories.

remap_cat_dict = {
     'Care Home Needs Assessment & Personalised Care and Support Planning': 'Care Home',
     'Care Home Visit': 'Care Home',
     'Clinical Triage': 'Clinic',
     'General Consultation Acute': 'Consultation',
     'General Consultation Routine': 'Consultation',
     'Group Consultation and Group Education': 'Consultation',
     'Home Visit': 'Home Visit',
     'Inconsistent Mapping': 'Miscellaneous',
     'Non-contractual chargeable work': 'Miscellaneous',
     'Patient contact during Care Home Round': 'Care Home',
     'Planned Clinical Procedure': 'Clinic',
     'Planned Clinics': 'Clinic',
     'Service provided by organisation external to the practice': 'Miscellaneous',
     'Social Prescribing Service': 'Social Prescribing Service',
     'Structured Medication Review' : 'Structured Medication Review',
     'Unmapped': 'Miscellaneous',
     'Unplanned Clinical Activity': 'Clinic',
     'Walk-in': 'Walk-in'
}


new_cat = nc_ss['national_category']
unique_cats = sorted(nc_ss['national_category'].unique())

for cat in unique_cats:
    if cat in remap_cat_dict:
        new_cat = new_cat.replace(cat, remap_cat_dict[cat])
    
nc_ss['new_category'] = new_cat
```

```python
#View cleaned DF. 
nc_ss
```

```python
# New dataframe with amended categories and groupby function.
nc_nc_2 = nc_ss.groupby([nc_ss.new_category, nc_ss.appointment_month])['count_of_appointments'].sum().reset_index()
```

```python
# Replotted.
sns.lineplot(x='appointment_month', 
            y='count_of_appointments', 
            data=nc_nc_2, 
            hue='new_category', 
            ci=None)

plt.yscale('log')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.title('Comparison of Cleaned Context Types over Time')

plt.savefig("output.png")
```

### Would it be useful to only look at retweeted and favourite tweet messages?
#### Explain your answer.
Retweeted and favourited messages are certainly useful, as they are representative of the shared opinion of multiple people rather than just one, but the NHS would like to understand whether there has been a high volume of hashtags relating to healthcare, which is a quantitative question. While it is important to understand the sentiment in these tweets, it is also important to measure them. As such I will conduct analysis both on only retweeted/favourited tweets, and on all tweets.

```python
# Create a new DataFrame containing only the text.
tweets_text = tweets['tweet_full_text']

# View the DataFrame.
tweets_text
```

```python
# Create an empty variable.
tags = []

# Create a for loop, filtering out website URLs.
for tweet_list in [tweet.split(' ') for tweet in tweets['tweet_full_text'].values]:
    for word in tweet_list:
        if '#' in word and 'https' not in word:
            tags.append(word.lower())
            
# Create a PD series to count the values in the list.
tags_series = pd.Series(tags)
tags_counts = tags_series.value_counts()
```

```python
# Display the first 30 records.
tags_counts[:30]
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
data = pd.DataFrame(tags_counts).reset_index()

# Rename the columns.
data.columns = ['word', 'count']

# View output. 
data
```

```python
# Fix the count datatype.
data.dtypes
```

```python
# Display records where the count is larger than 10.
data_over_10 = data[data['count'] > 10]
data_over_10
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
sns.barplot(x='count', 
            y='word', 
            data=data_over_10,
            ci=None)
plt.title('Most Used Hashtags')
plt.savefig('tweets1.png')
```
